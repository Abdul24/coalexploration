```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
sns.set();
```

```python
df = pd.read_excel("../data/coalpublic2013.xls", header=2, index_col='MSHA ID')
```

```python
df.head()
```

```python
df['Company Type'].unique()
```

```python
df['Company Type'].replace(to_replace='Indepedent Producer Operator',
                          value='Independent Producer Operator',
                          inplace=True)

df.head()
```

```python
df['Company Type'].head()
```

```python
df.rename(columns= lambda x: x.replace(" ", "_"), inplace=True)
```

```python
df.head()
```

```python
len(df)
```

```python
plt.scatter(df.Average_Employees, df.Labor_Hours)
plt.xlabel("Number of Employees")
plt.ylabel("Total Hours Worked")
```

```python
# Linear Regression Plot 
sns.regplot(df.Average_Employees, df.Labor_Hours)
plt.savefig("../figures/2017-04-16" + "-employees_vs_hours.png")
```

```python
# For loop to extract out column names in dataframe of dataset
for column in df.columns:
    print (column)
```

```python
plt.scatter(df.Labor_Hours, df['Production_(short_tons)'])
```

```python
#Histogram 
df['Production_(short_tons)'].hist()
```

```python
#Transformations 
# Return rows where all production equal zero
df[df['Production_(short_tons)'] ==0]
```

```python
# We are now removing data here by slicing! Storing as new dataframe.
# All prodction greater-than zero
df = df[df['Production_(short_tons)'] > 0]
```

```python
len(df)
```

```python
# Taking the log of production using numpy 
# And creating a new column that contains the log values
# Target varaible is the log production of these mines
df['log_production'] = np.log(df['Production_(short_tons)'])
```

```python
df.log_production.hist()
#Results very close to a log normal distribution 
```

```python
#Saving this dataframe as csv. In the event its needed for any future analysis
df.to_csv("../data/cleaned_coalpublic2013.csv")

#Explortory analysis completed
```
